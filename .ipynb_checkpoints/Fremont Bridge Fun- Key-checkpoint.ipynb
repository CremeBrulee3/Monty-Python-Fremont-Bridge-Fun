Read in the csv from the appropriate directory as a pandas dataframe.

```python
data = pd.read_csv('FremontBridge.csv', index_col='Date', parse_dates=True)
```

Rename the columns to simpler (one word) names, replace any empty data with the value 0 and add a column for the total number of riders.

```python
data.columns = ['West', 'East']
data.fillna(0, inplace=True)
data['Total'] = data.eval('East + West')
```

How many total people cross the bridge each day? Make a graph that shows this. How many people on average only make a one-way trip?

```python
data.resample('W', how='sum').plot()
plt.ylabel('weekly trips');
```

How does the number of rides vary by day of the week? Make a graph that illustrates this trend.

```python
days = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
data['dayofweek'] = data['Total'].index.dayofweek

grouped = data.groupby('dayofweek')['Total'].mean()
grouped.index = days

grouped.plot()
plt.title("Average Traffic By Day")
plt.ylabel("Average Daily Crossings");
```
